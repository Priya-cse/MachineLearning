```python
df["User_ID"] = df["User_ID"].astype("string")
df["month"] = df["month"].astype("string")
```

```python
new = df["month"].copy()
df["User_ID"] = df["User_ID"].str.cat(new, sep ="_")
```

```python
df["year"] = df["year"].astype("string")
```

```python
new = df["year"].copy()
df["User_ID"] = df["User_ID"].str.cat(new, sep ="_")
```
